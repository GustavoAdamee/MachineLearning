<h2>Função para imprimir os resultados</h2>

```python
def imprimeResult(results):
    media = results['test_score'].mean()
    print("Resultados cross validation:",results['test_score'])
    print("Média da precisão: {0:.2f}".format(media*100),'%')

    #Acurracy fica entre a média + 2(desvio_padrao) e a média - 2(desvio_padrao)
    desvio_padrao = results['test_score'].std()
    print("Desvio padrão: {0:.2f}".format(desvio_padrao*100),'%')

    acumulo_baixo = media - 2*desvio_padrao
    acumulo_alto = media + 2*desvio_padrao
    print("Acurracy esta entre: [",acumulo_baixo*100,"---", acumulo_alto*100,"]")
```

<h3>A precisão é alta mas, é possível notar que o modelo está treinando com poucas informações de fraude. (UnderSampling)</h3><br>

<h1> É necessário realizar um treino com um campo de teste mais 'justo'</h1>
<p> Para isso, é necessário reduzir os casos 'não fraudados' para 492 também (50/50) </p> 
<p> Um problema é que o modelo irá treinar com menos dados </p>

```python
#Pegando uma amostra de dados para testar o modelo

amostra_dataFrame = dataFrame.sample(frac = 1)  #frac = 1 pega todos os dados

fraude = amostra_dataFrame[amostra_dataFrame.Class == 1]
n_fraude = amostra_dataFrame[amostra_dataFrame.Class == 0][:492] #amostra de 492 casos não fraudados para o treino do modelo
```

```python
#Juntando as amostras para o treino
dataFrame_justo = pd.concat([fraude, n_fraude])
dataFrame_justo
```

984 linhas, sendo metade delas fraude e a outra metade não fraudadas.

<h1>Linear SVC com dados justos</h1>

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import cross_validate
from sklearn.model_selection import StratifiedKFold

x = dataFrame_justo.drop('Class', axis = 'columns')
y = dataFrame_justo['Class']


raw_train_x, raw_test_x, train_y, test_y = train_test_split(x, y, test_size = 0.25) #stratified não é necessário

model = LinearSVC()
cv = StratifiedKFold(n_splits=5, shuffle=True)
results = cross_validate(model, raw_train_x, train_y, cv = cv, return_train_score = False)
imprimeResult(results)
```

<h1>SVC() com dados justos</h1>

```python
#Escalando as dimensões de análise do modelo usando pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_validate
from sklearn.model_selection import StratifiedKFold

SEED = 301
np.random.seed(SEED)

scaler = StandardScaler()
model = SVC()

pipeline = Pipeline([('scaler', scaler), ('model', model)])

cv = StratifiedKFold(n_splits = 5, shuffle = True)

#AO invés do modelo, aqui é passado o pipeline
result = cross_validate(pipeline, x, y, cv = cv)

imprimeResult(result)
```

<h1>DECISION TREE com dados justos</h1>

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_validate
from sklearn.model_selection import StratifiedKFold

model = DecisionTreeClassifier()

cv = StratifiedKFold(n_splits = 5, shuffle = True)

result = cross_validate(model, x, y, cv = cv)

imprimeResult(result)
```
