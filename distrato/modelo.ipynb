```python
import pandas as pd
import datetime
```

```python
def imprimeResult(results):
    media = results['test_score'].mean()
    print("Resultados cross validation:",results['test_score'])
    print("Média da precisão: {0:.2f}".format(media*100),'%')

    #Acurracy fica entre a média + 2(desvio_padrao) e a média - 2(desvio_padrao)
    desvio_padrao = results['test_score'].std()
    print("Desvio padrão: {0:.2f}".format(desvio_padrao*100),'%')

    acumulo_baixo = media - 2*desvio_padrao
    acumulo_alto = media + 2*desvio_padrao
    
```

```python
df = pd.read_csv('infos.csv')
```

```python
df
```

```python
#removendo NaN
#Aqui talvez seja melhor tratar os dados ao invés de removê-los
df.dropna(inplace=True)
df
```

```python
#Selecionando novos contratos para testar o modelo

year_ago = datetime.datetime.now() - datetime.timedelta(days=365)
df_novos_contratos = pd.DataFrame() 

df.data_contrato = pd.to_datetime(df.data_contrato)

df_novos_contratos = df[df.data_contrato > year_ago]
df.drop(df[df.data_contrato > year_ago].index, inplace=True)
    
```

```python
df_novos_contratos
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

x = df.drop(['status', 'id_contrato', 'data_contrato'], axis=1)

y = df['status']

print(y.value_counts())

SEED = 1231
np.random.seed(SEED)

raw_train_x, raw_test_x, train_y, test_y = train_test_split(x, y, test_size=0.25, stratify=y)
```

<h1>LINEAR</h1>

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import cross_validate
from sklearn.model_selection import StratifiedKFold

model = LinearSVC()

cv = StratifiedKFold(n_splits=5)

cv_results = cross_validate(model, raw_train_x, train_y, cv=cv, return_train_score=False)

imprimeResult(cv_results)
```

<h1>DUMMY CLASSIFIER</h1>

```python
from sklearn.dummy import DummyClassifier

dummy_stratified = DummyClassifier(strategy='stratified')
dummy_most_frequent = DummyClassifier(strategy='most_frequent')

dummy_stratified.fit(raw_train_x, train_y)
dummy_most_frequent.fit(raw_train_x, train_y)
print('Acurácia do modelo dummy stratified: {}%'.format(dummy_stratified.score(raw_test_x, test_y)*100))
print('Acurácia do modelo dummy most frequent: {0:.2f}%'.format(dummy_most_frequent.score(raw_test_x, test_y)*100))
```

<h1>DECISION TREE</h1>

```python
from sklearn.tree import DecisionTreeClassifier

model_decision_tree = DecisionTreeClassifier(max_depth=3)

cv = StratifiedKFold(n_splits=5)

cv_results = cross_validate(model_decision_tree, raw_train_x, train_y, cv=cv, return_train_score=False)

imprimeResult(cv_results)
```

<h1>SVC</h1>

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import cross_validate

SEED = 301
np.random.seed(SEED)

scaler = StandardScaler()
model = SVC()

pipeline = Pipeline([('scaler', scaler), ('model', model)])

cv = StratifiedKFold(n_splits=5)

#AO invés do modelo, aqui é passado o pipeline
result = cross_validate(pipeline, raw_train_x, train_y , cv=cv)

imprimeResult(result)
```
